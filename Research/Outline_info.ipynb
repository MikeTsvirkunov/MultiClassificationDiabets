```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
def box_plot_show(data, nr, nc, c):
    fig = plt.figure(tight_layout=True)
    fig.set_figheight(25)
    fig.set_figwidth(8)
    gs = gridspec.GridSpec(nr, nc)
    for i in range(nr):
        for j in range(nc):
            ax = fig.add_subplot(gs[i, j])
            ax.boxplot(data[c[i+j]])
            ax.set_title(c[i+j])
    fig.show()
```

```python
data = pd.read_csv('../DATA/diabetes_012_health_indicators_BRFSS2015.csv', index_col=False)
```

```python
box_plot_show(data, 8, 3, data.columns)
```
