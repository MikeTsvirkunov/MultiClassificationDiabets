```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../DATA/split_classes.csv', index_col=False)
x = data.drop(['diabets', 'pre_diabets', 'non_diabets'], axis=1)
diabets = data['diabets']
pre_diabets = data['pre_diabets']
non_diabets = data['non_diabets']
```

```python
print('diabets',
      pd.value_counts(diabets),
      '\npre_diabets',
      pd.value_counts(pre_diabets),
      '\nnon_diabets',
      pd.value_counts(non_diabets),
      sep='\n')
```

```python
diabets_true_index = np.where(diabets == 1)[0]
diabets_false_index = np.where(diabets == 0)[0]
np.random.shuffle(diabets_false_index)
pre_diabets_true_index = np.where(pre_diabets == 1)[0]
pre_diabets_false_index = np.where(pre_diabets == 0)[0]
np.random.shuffle(pre_diabets_false_index)
non_diabets_true_index = np.where(non_diabets == 1)[0]
non_diabets_false_index = np.where(non_diabets == 0)[0]
np.random.shuffle(non_diabets_false_index)
```

```python
diabets_data = data.drop(['pre_diabets', 'non_diabets'], axis=1).iloc[diabets_true_index]
diabets_data = pd.concat([diabets_data, 
                          data.drop(['pre_diabets', 
                                     'non_diabets'], 
                                    axis=1).iloc[diabets_false_index[0:pd.value_counts(diabets).min()]]],
                          axis=0)
diabets_data
```

```python
diabets_data.sample(frac=1).reset_index(drop=True).to_csv('../DATA/diabets_classes.csv', index=False)
```

```python
pre_diabets_data = data.drop(['diabets', 'non_diabets'],
                         axis=1).iloc[pre_diabets_true_index]
pre_diabets_data = pd.concat([pre_diabets_data,
                          data.drop(['diabets',
                                     'non_diabets'],
                                    axis=1).iloc[pre_diabets_false_index[0:pd.value_counts(pre_diabets).min()]]],
                         axis=0)
pre_diabets_data
```

```python
pre_diabets_data.sample(frac=1).reset_index(drop=True).to_csv('../DATA/pre_diabets_classes.csv', index=False)
```

```python
non_diabets_data = data.drop(['diabets', 'pre_diabets'],
                             axis=1).iloc[non_diabets_true_index]
non_diabets_data = pd.concat([non_diabets_data,
                              data.drop(['diabets',
                                         'pre_diabets'],
                                        axis=1).iloc[non_diabets_false_index[0:pd.value_counts(non_diabets).min()]]],
                             axis=0)
non_diabets_data
```

```python
non_diabets_data.sample(frac=1).reset_index(drop=True).to_csv('../DATA/non_diabets_classes.csv', index=False)
```

# All

```python
diabets_all_data = data.drop(['pre_diabets', 'non_diabets'], axis=1).iloc[diabets_true_index]
diabets_all_data.rename(columns={'diabets': 'diabets'})
diabets_all_data['diabets'] = diabets_all_data['diabets'] * 2
pre_diabets_all_data = data.drop(['diabets', 'non_diabets'], axis=1).iloc[pre_diabets_true_index]
pre_diabets_all_data = pre_diabets_all_data.rename(columns={'pre_diabets': 'diabets'})[0:len(diabets_true_index)]
pre_diabets_all_data['diabets'] = pre_diabets_all_data['diabets'] * 1
non_diabets_all_data = data.drop(['pre_diabets', 'diabets'], axis=1).iloc[non_diabets_true_index]
non_diabets_all_data = non_diabets_all_data.rename(columns={'non_diabets': 'diabets'})[0:len(diabets_true_index)]
non_diabets_all_data['diabets'] = non_diabets_all_data['diabets'] * 0
all = pd.concat([diabets_all_data, pre_diabets_all_data, non_diabets_all_data], axis=0).sample(frac=1).reset_index(drop=True)
all
```

```python
all.to_csv('../DATA/down_sampling.csv', index=False)
```
