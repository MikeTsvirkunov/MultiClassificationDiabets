```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('../DATA/regression_remove_data.csv')
data
```

```python
X, y = data.drop('Diabetes_012', axis=1), data['Diabetes_012']
lab = LabelEncoder()
y = lab.fit_transform(y)
data_balanced, y_balanced = SMOTE().fit_resample(X, y)
data_balanced['Diabetes_012'] = y_balanced
```

```python
cs = OneHotEncoder(sparse_output=False).fit_transform(data_balanced['Diabetes_012'].to_numpy().T.reshape(-1, 1))
cs = pd.DataFrame(cs, columns=['non_diabets', 'diabets', 'pre_diabets'])
cs
```

```python
pd.concat([data_balanced.drop('Diabetes_012', axis=1), cs], axis=1).to_csv('../DATA/split_classes.csv', index=False)
```
