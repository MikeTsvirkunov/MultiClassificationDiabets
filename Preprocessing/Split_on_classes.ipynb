```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv('../DATA/regression_remove_data.csv')
data
```

```python
cs = OneHotEncoder(sparse_output=False).fit_transform(data['Diabetes_012'].to_numpy().T.reshape(-1, 1))
cs = pd.DataFrame(cs, columns=['non_diabets', 'diabets', 'pre_diabets'])
cs
```

```python
pd.concat([data.drop('Diabetes_012', axis=1), cs], axis=1).to_csv('../DATA/split_classes.csv', index=False)
```
